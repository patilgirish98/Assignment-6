2. How does broadcasting work in NumPy ?

    - Broadcasting in NumPy allows operations between arrays of different shapes without explicitly replicating data, making computations memory-efficient and fast.
    - When performing operations (e.g., addition, subtraction) on arrays of different shapes, NumPy automatically expands the smaller array to match the shape of the larger array without copying data.
    - Example :-

          import numpy as np

          A = np.array([[1, 2, 3], [4, 5, 6]])
          B = np.array([10, 20, 30])            

          print(A + B)  # B is broadcasted to (2,3)

17. Explain the difference between apply() and map() in Pandas ?

        - Works on DataFrames and Series.

    - map() :-

        - Works only on Series.
        - Primarily used for applying a function to each element in a single Series.
        - Supports element-wise transformations and can map values using dictionaries or Series

21. Why is NumPy’s array slicing faster than Python’s list slicing ?

    - NumPys slicing is faster because it uses contiguous memory and creates views instead of copies, with optimized operations in C.
    - Python list slicing is slower due to copying data and higher overhead from Python's dynamic nature.
       - Memory Efficiency: NumPy arrays are stored in contiguous memory, making slicing faster, while Python lists store references to objects, adding overhead.
       - View vs Copy: NumPy slicing returns a view (no data copying), whereas Python list slicing creates a new list
       - Optimized C Backend: NumPy slicing is implemented in compiled C code, making it faster than Python's list slicing, which is managed by the interpreter.
   - In summary, NumPy is more memory-efficient and optimized for speed compared to Python lists

```python
# How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [1, 3, 5, 7, 9],
    'D': [9, 7, 5, 3, 1]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Matrix Heatmap')
plt.show()
```
